## 1 n皇后

最常见的八皇后问题，在国际象棋棋盘上摆放8个皇后，要求：每个皇后都不能吃到其他皇后。（每个横排、竖排、45度斜线方向只能有一个皇后）



正向思路：

从第一排开始，逐排填皇后，每次选择该行靠左的位置来填，某行不能填了则向上一行回退

```python
N = 8
Q = []
# Q列表保存每行位置，-1代表没有填
for i in range(N):
    Q.append(-1)

print(Q)
```

皇后的坐标位置

```python
for j in range(N):
    for i in range(N):
        print("(%d,%d)"%(j,i), end = " ")
    print("")
```

可以看出对角线皇后坐标的规律，左上到右下方向i-j相等，右上到左下方向i+j相等。

判断皇后是否可放

```python
def isValid(row, col):
    Flag = True
    # 自上而下按照行来放置皇后
    # 首先判断是否在上面行的相同列上有皇后
    for i in range row:
        if(col==Q[row]):
            return False
    
```
